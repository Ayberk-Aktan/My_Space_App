<h4>Astreoids Risk Analysis Dashboard (ARAD) Risk Model Notebook<h4> 
<p>This notebook will be used to test a Random Forrest model.</p>

```python
import os 
import sqlite3 
import pandas as pd 
```

```python
db_path = os.path.join("..","astreoid_risk_anaylzed.db")
conn = sqlite3.connect(db_path) 
df = pd.read_sql("SELECT * FROM astreoids WHERE close_date",conn) 
```

```python
df.head(3)
```

```python
df.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report , confusion_matrix 
from sklearn.model_selection import train_test_split 
```

```python
X = df.drop(["close_date","astreoid_name","min_distance_km","potentially_hazardous"],axis=1)  
Y = df["potentially_hazardous"] 
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size=0.3,random_state=22)
```

```python
forest = RandomForestClassifier(n_estimators=200,class_weight="balanced",max_depth=5,min_samples_leaf=5,random_state=22) 
model = forest.fit(x_train,y_train) 

predict = model.predict(x_test)

print(model.score(x_test,y_test))
print(model.score(x_train,y_train))
```

```python
print(classification_report(y_test,predict))
```

```python
confusion_matrix(y_test,predict)
```

```python
"""
! Google Gemini was used to export the model in .pkl format.
"""
import joblib 

joblib.dump(forest,"model.pkl")
```
